```python
numerical_features = [feature for feature in df.columns if df[feature].dtypes != 'O']

print('Number of numrical variables: ', len(numerical_features))

#visualize the numrical variables
df[numerical_features].head()
```

```python
year_features = [feature for feature in numerical_features if 'month' in feature or 'year' in feature]

print(year_features)
```

```python
for feature in year_features:
        print(feature,df[feature].unique())
```

```python
# Step 5: Descriptive statistics
# Convert percentages to proper scales
df['danceability_%'] = df['danceability_%'] / 100
df['valence_%'] = df['valence_%'] / 100
df['energy_%'] = df['energy_%'] / 100
df['acousticness_%'] = df['acousticness_%'] / 100
df['liveness_%'] = df['liveness_%'] / 100
df['speechiness_%'] = df['speechiness_%'] / 100
```

```python
# Correlation heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(df[numerical_features].corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Ensure the streams column is numeric
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
```

```python
year_features = [feature for feature in numerical_features if 'year' in feature]

print(year_features)
```

```python
# Let's explore the content of these year information.
for feature in year_features:
        print(feature,df[feature].unique())
```

```python
df.groupby('released_year')['streams'].median().plot()
plt.xlabel('releaseyear')
plt.ylabel('Median song stream')
plt.title("stream vs release")
```

```python
# Distribution of BPM
plt.figure(figsize=(15, 10))
sns.histplot(df['bpm'], kde=True)
plt.title('Distribution of BPM')
plt.show()
```

```python
#insight-Moderato – moderately (98–112 BPM) Allegretto – by the mid-19th century, moderately fast (102–110 BPM) Allegro – fast, quickly and bright (120–156 BPM) (molto allegro is slightly faster than allegro, but always in its range; 124-156 BPM). Vivace – lively and fast (156–176 BPM)
```
